```python
from sklearn.preprocessing import LabelEncoder
for column in data.columns:
        if data[column].dtype == np.number:
            continue
        data[column] = LabelEncoder().fit_transform(data[column])
data.head()
```

```python
#Split the data into independent 'X' and dependent 'Y' variables
X = data.iloc[:, 1:data.shape[1]].values 
Y = data.iloc[:, 0].values 
```

```python
X  = data.drop(['Yes'], axis = 1)
Y = data['Yes']
```

```python
#split data into 70% training and 30% testing.
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
```

```python
RF.fit(X_train,Y_train)
```

```python
#Know the accuracy on the training data set
RF.score(X_train,Y_train)
```

```python
#Confusion Matrix and score of the model on the test data
from sklearn.metrics import confusion_matrix
ConMatrix = confusion_matrix(Y_test,RF.predict(X_test))

TN = ConMatrix[0][0]
TP = ConMatrix[1][1]
FN = ConMatrix[1][0]
FP = ConMatrix[0][1]

print(ConMatrix)
print('The model testing accuracy = {}'.format((TP+TN)/(TP+TN+FN+FP)))
```

```python
# Returns the feature importances (Higher value indicates the  feature is more important).
imp = pd.DataFrame({'feature':data.iloc[:, 1:data.shape[1]].columns,'importance':np.round(RF.feature_importances_,3)}) 
#The target column is at position 0
imp = imp.sort_values('importance',ascending=False).set_index('feature')
imp
```
